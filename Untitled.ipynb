```python
import numpy as np
from scipy import stats

def kaplan_meier(T, E, alpha=0.05):
    T, E = np.asarray(T, float), np.asarray(E, int)
    event_times = np.unique(T[E == 1])

    S = [1.0]
    var_sum = [0.0]
    at_risk, events = [], []

    for t in event_times:
        n = np.sum(T >= t)
        d = np.sum((T == t) & (E == 1))

        S.append(S[-1] * (1 - d / n))
        var_sum.append(var_sum[-1] + (d / (n * (n - d)) if n > d else 0))

        at_risk.append(n)
        events.append(d)

    S = np.array(S)
    var_sum = np.array(var_sum)
    times = np.concatenate(([0], event_times))

    # Confidence intervals (log-log)
    z = stats.norm.ppf(1 - alpha / 2)
    logS = np.log(-np.log(S[1:] + 1e-15))
    se = z * np.sqrt(var_sum[1:]) / np.abs(np.log(S[1:] + 1e-15))

    lower = np.exp(-np.exp(logS + se))
    upper = np.exp(-np.exp(logS - se))

    lower = np.concatenate(([1.0], np.clip(lower, 0, 1)))
    upper = np.concatenate(([1.0], np.clip(upper, 0, 1)))

    return {
        "times": times,
        "survival": S,
        "lower": lower,
        "upper": upper,
        "at_risk": [len(T)] + at_risk,
        "events": [0] + events
    }


def km_median(km):
    idx = np.where(km["survival"] <= 0.5)[0]
    return km["times"][idx[0]] if len(idx) else np.nan
```
